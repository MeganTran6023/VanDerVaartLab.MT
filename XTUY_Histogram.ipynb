## 1- Import CSV 

```python
#All data

import pandas as pd

input_file = pd.read_csv("XTUY_Histogram.csv")

#removes sequences without any register shifts
input_file.dropna()
```

```python
#Uracil Major Flip

input_major = pd.read_csv("XTUY_major.csv")

#removes sequences without any register shifts
input_major.dropna()
```

```python
#Uracil Minor Flip

input_minor = pd.read_csv("XTUY_minor.csv")

#removes sequences without any register shifts
input_minor.dropna()
```

## 2- Make Histogram

### 2a) Histogram for register shift (RS) duration

```python
import matplotlib.pyplot as plt

input_file['How long RS last? (frames)'].plot(kind='hist')

#labels
plt.title("XTUY Sequences: Frequency of Register Shifts")
plt.xlabel("Register Shift Duration (frames)")
```

### 2b) Histogram Major flip

```python
import matplotlib.pyplot as plt

input_major['How long RS last? (frames)'].plot(kind='hist',bins=15)

#labels
plt.title("XTUY Major Sequences: Frequency of Register Shifts")
plt.xlabel("Register Shift Duration (frames)")
```

### 2c) Histogram Minor Flip

```python
import matplotlib.pyplot as plt

input_minor['How long RS last? (frames)'].plot(kind='hist',bins=15)

#labels
plt.title("XTUY Minor Sequences: Frequency of Register Shifts")
plt.xlabel("Register Shift Duration (frames)")
```
