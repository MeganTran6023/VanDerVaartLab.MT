## 1- Import Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 2- Form Bar Graphs

```python
#Average RMSD
x_1 = ["Block 1","Block 2"]
y_1 = [1.2172,1.06294]

plt.bar(x_1,y_1)

##add actual values above respective bars
def addLabels(x_1,y_1):
    for i in range(len(x_1)):
        plt.text(i,y_1[i],y_1[i], ha="center")

addLabels(x_1,y_1)

plt.bar(x_1,y_1,color = "salmon")

##labels
plt.title("RMSD Averages")
plt.ylabel("RMSD (Å)")
```

```python
#Median RMSD
x_2 = ["Block 1","Block 2"]
y_2 = [1.3694,0.54685]

plt.bar(x_2,y_2)

##add actual values above respective bars
def addLabels(x_2,y_2):
    for i in range(len(x_2)):
        plt.text(i,y_2[i],y_2[i], ha="center")

addLabels(x_2,y_2)

plt.bar(x_2,y_2,color = "salmon")

##labels
plt.title("RMSD Medians")
plt.ylabel("RMSD (Å)")
```
