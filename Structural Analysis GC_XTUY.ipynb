# Bar Graph

a) Rise Major

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame(new_rise_major)

X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])

# Plot the data using bar() method
plt.bar(X, Y, color='g')
plt.title("Average Rise XTUY Major Sequences")
plt.xlabel("Sequences")
plt.ylabel("Degrees")

# Change y-axis limits for better visualization
plt.ylim(0, max(Y) * 1.5)

# Rotate x-labels by 45 degrees
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()  
plt.show()
```
