```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
```

## 1- Import CSV

```python
#Uracil Major Flip

input_major = pd.read_csv("XTUY_major.csv")

#removes sequences without any register shifts
input_major.dropna()
```

```python
#Uracil Minor Flip

input_minor = pd.read_csv("XTUY_minor.csv")

#removes sequences without any register shifts
input_minor.dropna()
```

## 2- Bar Graph

One graph for major flip, another for minor flip

```python
#sum up duration for major RS
major_sum = input_major["How long RS last? (frames)"].sum()

major_sum
```

```python
#sum up duration for minor RS
minor_sum = input_minor["How long RS last? (frames)"].sum()

minor_sum
```

```python
#make bar graph
# x = major and minor
# y = duration total

x = ["Major","Minor"]
y = [38095,118642]

plt.bar(x,y)

#boundaries y-axis
plt.ylim(bottom=0, top=150000)
plt.yticks([25000,50000,75000,100000,125000,150000])

#bar graph
plt.bar(x,y, color = "salmon")


#labels
plt.title("Total Duration of Register Shifts")
plt.xlabel("Type of Uracil Flip")
plt.ylabel("Duration (frames)")
```
