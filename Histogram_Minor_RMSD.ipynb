## 1- Import Libraries

```python
import pandas as pd
import numpy as np
```

## 2- Opening .txt files of RMSD data Major Sequences

```python
#Block 1 
rmsd_data_b1 = pd.read_csv("rmsd_avg_block1_minor.txt", sep=" ", header=None,names=["Sequence", "RMSD Average"])

rmsd_data_b1
```

```python
#fix txt file

#1- Remove specific string 'XTUY'
strings_to_remove = ['XTUY/', ':','minor','_']

#2- loop through each row to see if XTUY and : needs to be removed and does it
for string in strings_to_remove:
    rmsd_data_b1['Sequence'] = rmsd_data_b1['Sequence'].str.replace(string, '')

#3- make new edited file
rmsd_data_b1.to_csv("rmsd_avg_block1_minor.txt", sep=' ', index=False)

rmsd_data_b1
```

```python
#Block 2 
rmsd_data_b2 = pd.read_csv("rmsd_avg_block2_minor.txt", sep=" ", header=None,names=["Sequence", "RMSD Average"])

rmsd_data_b2
```

```python
#fix txt file

#1- Remove specific string 'XTUY'
strings_to_remove = ['XTUY/', ':', '_','minor']

#2- loop through each row to see if XTUY and : needs to be removed and does it
for string in strings_to_remove:
    rmsd_data_b2['Sequence'] = rmsd_data_b2['Sequence'].str.replace(string, '')

#3- make new edited file
rmsd_data_b2.to_csv("rmsd_avg_block1_major.txt", sep=' ', index=False)

rmsd_data_b2
```
