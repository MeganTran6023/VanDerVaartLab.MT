## 2) Change coordinates so base flips

```python
import pandas as pd

def flip_molecule_axis(input_file, output_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(input_file)
    
    # Flip the X Y and Z-coordinates by negating them
    df['X'] = -df['X']
    df['Y'] = -df['Y']
    df['Z'] = -df['Z']
    
    # Write the modified DataFrame back to a new CSV file
    df.to_csv(output_file, index=False)

# Usage example
input_file = "coordinates-resid9.csv"
output_file = "coordinates-resid9_output.csv"
flip_molecule_axis(input_file, output_file)
```
